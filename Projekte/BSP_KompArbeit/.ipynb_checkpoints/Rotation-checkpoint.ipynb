```python
# for plotting the Picture
def printarray(incoming):
    LengthLevel1 = len(incoming) # nur die Anzahl der Zeilen
    LengthLevel2 = len(incoming[1]) # nur die Anzahl der spalten
    emptyString = ""
    for i in range(0, LengthLevel1,1): # Zeile
        for j in range(0, LengthLevel2,1): #Spalten
            if(incoming[i][j] == 0):
                emptyString = emptyString + "| |"
            elif(incoming[i][j] == 1):
                emptyString = emptyString + "|#|"
            elif(incoming[i][j] == 2):
                emptyString = emptyString + "|~|"
            elif(incoming[i][j] == 3):
                emptyString = emptyString + "|%|"
            elif(incoming[i][j] == 4):
                emptyString = emptyString + "|$|"
            elif(incoming[i][j] == 5):
                emptyString = emptyString + "|&|"
            else:
                pass
        emptyString = emptyString + "\n"
    print(emptyString)
                   
            
            
        
    
```

```python
# hier der Schalter zwischen den Testfiles
workingfile  = Testfile2
```

```python
def rotatePic(incoming):
    returnValue = [] # die neu erstellte Matrix
    orginalLineLength = len(incoming)
    orginalColumnLength = len(incoming[0])
    # Transponieren in die neuen wer
    newLineLength = orginalColumnLength
    newColumnLength = orginalLineLength
    
    valueForReading = orginalLineLength -1 # braucht es fr das auslesen aus der alten Matrix und schreiben in die neue
    for i in range(0,newLineLength,1): # loop ber die anzahl der neuen Zielen
        interimLineVector = [] # fr jede Zeile eigenen interimVector, jedes mal neu
        for j in range(0, newColumnLength,1): # loop ber die neuen Zeilen
            
            interimLineVector.append(incoming[valueForReading-j][i])
            '''
            Schreibe chronoloisch in die erste Zeile des neuen Matrix, 
            die werte der esrten Spalte, beginnend bei der Grten zeilenZahl Daher oben die in eigene Variable
            '''
        
        returnValue.append(interimLineVector)
    return(returnValue)
        
        
        
        
```

```python
rotatedPic = rotatePic(workingfile)
```
