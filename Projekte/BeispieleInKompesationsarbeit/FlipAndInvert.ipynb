```python
# for plotting the Picture
def printarray(incoming):
    LengthLevel1 = len(incoming) # nur die Anzahl der Zeilen
    LengthLevel2 = len(incoming[1]) # nur die Anzahl der spalten
    emptyString = ""
    for i in range(0, LengthLevel1,1): # Zeile
        for j in range(0, LengthLevel2,1): #Spalten
            if(incoming[i][j] == 0):
                emptyString = emptyString + "| |"
            elif(incoming[i][j] == 1):
                emptyString = emptyString + "|#|"
            elif(incoming[i][j] == 2):
                emptyString = emptyString + "|~|"
            elif(incoming[i][j] == 3):
                emptyString = emptyString + "|%|"
            elif(incoming[i][j] == 4):
                emptyString = emptyString + "|$|"
            elif(incoming[i][j] == 5):
                emptyString = emptyString + "|&|"
            else:
                pass
        emptyString = emptyString + "\n"
    print(emptyString)
```

```python
# annahme 0 bzw leer = zeigt den Green-Screen, 1 zeigt content
basicPicture = [[1,0],
                [0,1]             
                ]
printarray(basicPicture)
```

```python
basicarray3= [[0,1,1,1,0], 
              [1,0,0,0,1],
              [0,1,0,1,0], 
              [0,0,1,0,0]
]
printarray(basicarray3)
```

```python
# festlegen der working-Array
workingarray = basicarray3
#workingarray = basicPicture
```

```python
def switchTurn(incoming):
    returnValue = []
    for i in range(1,len(incoming)+1,1):
        returnValue.append(incoming[-i])
    return(returnValue)
```

```python
def invertPicture(incoming):
    returnValue = [] # returnvalue
    for i in range(0,len(incoming),1): # hier die zeilen
        interimVector = []# wird fr jede Zeile neu gebaut
        for j in range(0, len(incoming[i]),1):# iteriere durch die Spalten
            if((incoming[i][j]) == 1): # prfe ob 1 oder 0 und schreibe das gegenteil in den interim zeilen vektor || => invertiere
                interimVector.append(0)
            else:
                interimVector.append(1)
        returnValue.append(interimVector) # schreibe interim Zielenvektor in returnvalue und baue ggf den nchsten
    return(returnValue) # gibt den value zurck zum Methoden aufruf
        
```

```python
flipedPicture = switchTurn(workingarray)
invertedPicture = invertPicture(flipedPicture) # take Flipped Picture from step before 
```

```python
printarray(workingarray)
printarray(flipedPicture)
printarray(invertedPicture)
```
